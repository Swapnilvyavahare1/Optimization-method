```python
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
x_train
```

```python
x_train= x_train.reshape(x_train.shape[0], 28,28,1)
x_test= x_test.reshape(x_test.shape[0], 28,28,1)
input_shape= (28,28,1)
y_train= keras.utils.to_categorical(y_train)
y_test= keras.utils.to_categorical(y_test)
x_train= x_train.astype('float')
x_test= x_test.astype('float')
```

```python
x_train
```

Pre-processing step

Min-max approach

```python
x_train /=255
x_test /=255
```

Building DNN Model with all Optimization methods

```python
epochs= 10
batch_size= 64
num_classes= 10 #(0 to 9)

def build_model(optimizer):
    model= Sequential()
    model.add(Flatten())# This is mandatory for Image processing problem
    model.add(Dense(256, activation= 'relu', input_shape= input_shape))
    model.add(Dropout(0.25))
    model.add(Dense(256, activation= 'relu'))
    model.add(Dropout(0.25))
    model.add(Dense(num_classes, activation= 'softmax'))
    model.compile(optimizer= optimizer, loss= keras.losses.categorical_crossentropy, metrics= ['accuracy'])
    return model
```

Validate which optimization will work better?

```python
optimizer= ['SGD', 'Adagrad', 'Adadelta', 'RMSprop', 'Adam']
for i in optimizer:
    model= build_model(i)
    history= model.fit(x_train, y_train, validation_data= (x_test, y_test), batch_size= 64,
                       epochs= 10, verbose=1)
```

```python
#Conclusion
#'SGD' loss: 0.2091 - accuracy: 0.9396 - val_loss: 0.1551 - val_accuracy: 0.9541
#'Adagrad' accuracy: 0.8835 - val_loss: 0.2972 - val_accuracy: 0.9157
#'Adadelta' accuracy: 0.6282 - val_loss: 1.3002 - val_accuracy: 0.7460
#'RMSprop' accuracy: 0.9844 - val_loss: 0.0752 - val_accuracy: 0.9811
#'Adam' accuracy: 0.9866 - val_loss: 0.0631 - val_accuracy: 0.9826
```
